```python
%matplotlib inline

# load necessary libraries
import matplotlib.pyplot as plt
import numpy as np
from IPython.html.widgets import *
```

```python
# 1. slope intercept line
def f(x, m, b):
    return m*x+b

def draw_plot(m, b):
    x = np.linspace(0,1,100)
    y = np.array(map(lambda elem : f(elem, m, b), x))
    plt.plot(x, y)
    plt.show()
    
interact(draw_plot, m=(1.), b=(1.))
```

```python
# 2. point slope line
def f(x, m, x1, y1):
    return m*(x-x1)+y1

def draw_plot(m, x1, y1):
    x = np.linspace(0,1,100)
    y = np.array(map(lambda elem : f(elem, m, x1, y1), x))
    plt.plot(x, y)
    plt.show()
    
interact(draw_plot, m=(1.), x1=(1.), y1=(1.))
```

```python
# 3. standard line
def f(x, A, B, C):
    return 1.*(C-A*x)/B

def draw_plot(A, B, C):
    x = np.linspace(0,1,100)
    y = np.array(map(lambda elem : f(elem, A, B, C), x))
    plt.plot(x, y)
    plt.show()
    
interact(draw_plot, A=(1.), B=(1.), C=(1.))
```

```python
# 4. parabola vertex
def f(x, a, h, k):
    return a*(x-h)**2+k

def draw_plot(a, h, k):
    x = np.linspace(0,1,100)
    y = np.array(map(lambda elem : f(elem, a, h, k), x))
    plt.plot(x, y)
    plt.show()
    
interact(draw_plot, a=(1.), h=(.5), k=(1.))
```

```python
# 5. parabola standard form
def f(x, a, b, c):
    return a*x**2+b*x+c 

def draw_plot(a, b, c):
    x = np.linspace(0,1,100)
    y = np.array(map(lambda elem : f(elem, a, b, c), x))
    plt.plot(x, y)
    plt.show()
    
interact(draw_plot, a=(1.), b=(-1.), c=(1.))
```

```python
# 6. parabola horizontal form
def f(y, a, h, k):
    return a*(y-h)**2+k

def draw_plot(a, h, k):
    y = np.linspace(0,1,100)
    x = np.array(map(lambda elem : f(elem, a, h, k), y))
    plt.plot(x, y)
    plt.show()
    
interact(draw_plot, a=(1.), h=(.5), k=(1.))
```

```python
# 7. hyperbola horizontal form
def draw_plot(a, b, h, k):
    x = np.linspace(-9, 9, 400)
    y = np.linspace(-5, 5, 400)
    x, y = np.meshgrid(x, y)
    plt.contour(x, y,((x-h)**2/a**2 - (y-k)**2/b**2), [1])
    plt.show()
    
interact(draw_plot, a=(1.), b=(1.), h=(-2.,2.), k=(-2.,2.))
```

```python
# 8. hyperbola vertical form
def draw_plot(a, b, h, k):
    x = np.linspace(-9, 9, 400)
    y = np.linspace(-5, 5, 400)
    x, y = np.meshgrid(x, y)
    plt.contour(x, y,((y-k)**2/b**2 -(x-h)**2/a**2), [1])
    plt.show()
    
interact(draw_plot, a=(1.), b=(1.), h=(-2.,2.), k=(-2.,2.))
```

```python
# 9. circle standard form
def draw_plot(R, h, k):
    x = np.linspace(-5, 5, 400)
    y = np.linspace(-5, 5, 400)
    x, y = np.meshgrid(x, y)
    plt.contour(x, y,((y-k)**2/R**2 +(x-h)**2/R**2), [1])
    plt.axis('equal')
    plt.show()
    
interact(draw_plot, R=(2.),  h=(-2.,2.), k=(-2.,2.))
```

```python
# 10. ellipse standard form
def draw_plot(a, b, h, k):
    x = np.linspace(-5, 5, 400)
    y = np.linspace(-5, 5, 400)
    x, y = np.meshgrid(x, y)
    plt.contour(x, y,((y-k)**2/b**2 +(x-h)**2/a**2), [1])
    plt.axis('equal')
    plt.show()
    
interact(draw_plot, a=(2.), b=(1.), h=(-2.,2.), k=(-2.,2.))
```

```python
# 11. cosine standard form
def f(x, a, h, k):
    return a*np.cos(x-h) + k

def draw_plot(a, h, k):
    x = np.linspace(-5, 5, 400)
    y = np.array(map(lambda elem : f(elem, a, h, k), x))
    plt.plot(x, y)
    plt.show()
    
interact(draw_plot, a=(1.), h=(-1.), k=(1.))
```

```python
# 12. sine standard form
def f(x, a, h, k):
    return a*np.sin(x-h) + k

def draw_plot(a, h, k):
    x = np.linspace(-5, 5, 400)
    y = np.array(map(lambda elem : f(elem, a, h, k), x))
    plt.plot(x, y)
    plt.show()
    
interact(draw_plot, a=(1.), h=(-1.), k=(1.))
```

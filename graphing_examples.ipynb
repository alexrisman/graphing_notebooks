```python
%matplotlib inline

# load necessary libraries
import matplotlib.pyplot as plt
import numpy as np
from IPython.html.widgets import *
from collections import OrderedDict
```

```python
# 1. slope intercept line
def f(x, m, b):
    return m*x+b

def draw_plot(m, b):
    x_lim = 5
    x = list(np.linspace(-x_lim, x_lim, 400))
    y = list((map(lambda elem : f(elem, m, b), x)))
    plt.plot(x, y)
    plt.xlim(-x_lim, x_lim)
    plt.show()
    
interact(draw_plot, m=(1.), b=(1.))
```

```python
# 2. point slope line
def f(x, m, x1, y1):
    return m*(x-x1)+y1

def draw_plot(m, x1, y1):
    x_lim = 5
    x = list(np.linspace(-x_lim, x_lim, 400))
    y = list(map(lambda elem : f(elem, m, x1, y1), x))
    plt.plot(x, y)
    plt.xlim(-x_lim, x_lim)
    plt.show()
    
interact(draw_plot, m=(1.), x1=(1.), y1=(1.))
```

```python
# 3. standard line
def f(x, A, B, C):
    return 1.*(C-A*x)/B

def draw_plot(A, B, C):
    x_lim = 5
    x = list(np.linspace(-x_lim, x_lim, 400))
    y = list(map(lambda elem : f(elem, A, B, C), x))
    plt.plot(x, y)
    plt.show()
    
interact(draw_plot, A=(1.), B=(-1., 3., .3), C=(1.))
```

```python
# 4. parabola vertex
def f(x, a, h, k):
    return a*(x-h)**2+k

def draw_plot(a, h, k):
    x_lim = 5
    x = np.linspace(-x_lim, x_lim, 400)
    y = list(map(lambda elem : f(elem, a, h, k), x))
    plt.plot(x, y)
    plt.plot(h, k, 'ro', label='vertex')
    plt.plot(h, k+1./(4*a), 'ko', label='focus')
    print(k+1./(4*a))
    plt.legend(numpoints=1)
    plt.xlim(-x_lim, x_lim)
    plt.show()
    
interact(draw_plot, a=(-.5, 1.5, .3), h=(-1.,1.), k=(-1.,1.))
```

```python
# 5. parabola standard form
def f(x, a, b, c):
    return a*x**2+b*x+c 

def draw_plot(a, b, c):
    x_lim = 5
    x = np.linspace(-x_lim, x_lim, 400)
    y = list(map(lambda elem : f(elem, a, b, c), x))
    plt.plot(x, y)
    h = -b/(2*a)
    k = f(h, a, b, c)
    plt.plot(h, k, 'ro', label='vertex')
    plt.plot(h, k+1./(4*a), 'ko', label='focus')
    plt.legend(numpoints=1)
    plt.xlim(-x_lim, x_lim)
    plt.show()
    
interact(draw_plot, a=(-.5, 1.5, .3), b=(-1.,1.), c=(-1.,1.))
```

```python
# 6. parabola horizontal form
def f(y, a, h, k):
    return a*(y-h)**2+k

def draw_plot(a, h, k):
    y_lim = 5
    y = np.linspace(-y_lim, y_lim, 400)
    x = list(map(lambda elem : f(elem, a, h, k), y))
    plt.plot(k, h, 'ro', label='vertex')
    plt.plot(k+1./(4*a), h, 'ko', label='focus')
    plt.legend(numpoints=1)    
    plt.plot(x, y)
    plt.ylim(-y_lim, y_lim)
    plt.show()
    
interact(draw_plot, a=(-.5, 1.5, .3), h=(-1.,1.), k=(-1.,1.))
```

```python
# 7. hyperbola horizontal form
def draw_plot(a, b, h, k):
    axis_lim = 5
    x = np.linspace(-axis_lim, axis_lim, 400)
    y = np.linspace(-axis_lim, axis_lim, 400)
    asymptote_1 = k + 1.*b/a*(x - h)
    asymptote_2 = k - 1.*b/a*(x - h)
    plt.plot(list(x), list(asymptote_1), dashes=[3, 3], color='orange', label='asymptote')
    plt.plot(list(x), list(asymptote_2), dashes=[3, 3], color='orange', label='asymptote')
    x, y = np.meshgrid(x, y)
    plt.contour(x, y,((x-h)**2/a**2 - (y-k)**2/b**2), [1])
    plt.plot(h+a, k, 'ro', label='vertex')
    plt.plot(h-a, k, 'ro', label='vertex')
    plt.plot(h, k, 'go', label='center')
    c = (a**2+b**2)**.5
    plt.plot(h+c, k, 'ko', label='focus')
    plt.plot(h-c, k, 'ko', label='focus')
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = OrderedDict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys(), numpoints=1)
    plt.xlim(-axis_lim, axis_lim)
    plt.ylim(-axis_lim, axis_lim)
    plt.show()
    
interact(draw_plot, a=(-1., 3., .3), b=(-1., 3., .3), h=(-2.,2.), k=(-2.,2.))
```

```python
# 8. hyperbola vertical form
def draw_plot(a, b, h, k):
    axis_lim = 5
    x = np.linspace(-axis_lim, axis_lim, 400)
    y = np.linspace(-axis_lim, axis_lim, 400)
    asymptote_1 = h + 1.*a/b*(x - k)
    asymptote_2 = h - 1.*a/b*(x - k)
    plt.plot(list(x), list(asymptote_1), dashes=[3, 3], color='orange', label='asymptote')
    plt.plot(list(x), list(asymptote_2), dashes=[3, 3], color='orange', label='asymptote')
    x, y = np.meshgrid(x, y)
    plt.contour(x, y,((y-h)**2/a**2 -(x-k)**2/b**2), [1])
    plt.plot(k, h+a, 'ro', label='vertex')
    plt.plot(k, h-a, 'ro', label='vertex')
    plt.plot(k, h, 'go', label='center')
    c = (a**2+b**2)**.5
    plt.plot(k, h+c, 'ko', label='focus')
    plt.plot(k, h-c, 'ko', label='focus')
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = OrderedDict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys(), numpoints=1)
    plt.xlim(-axis_lim, axis_lim)
    plt.ylim(-axis_lim, axis_lim)
    plt.show()
    
interact(draw_plot, a=(-1., 3., .3), b=(-1., 3., .3), h=(-2.,2.), k=(-2.,2.))
```

```python
# 9. circle standard form
def draw_plot(R, h, k):
    axis_lim = 5
    x = np.linspace(-axis_lim, axis_lim, 400)
    y = np.linspace(-axis_lim, axis_lim, 400)
    x, y = np.meshgrid(x, y)
    plt.contour(x, y,((y-k)**2/R**2 +(x-h)**2/R**2), [1])
    plt.plot(h, k, 'go', label='center')
    plt.axis('equal')
    plt.xlim(-axis_lim, axis_lim)
    plt.ylim(-axis_lim, axis_lim)
    plt.show()
    
interact(draw_plot, R=(.1, 2.5, .3),  h=(-2.,2.), k=(-2.,2.))
```

```python
# 10. ellipse standard form
def draw_plot(a, b, h, k):
    axis_lim = 5
    x = np.linspace(-axis_lim, axis_lim, 400)
    y = np.linspace(-axis_lim, axis_lim, 400)
    x, y = np.meshgrid(x, y)
    plt.contour(x, y,((y-k)**2/b**2 +(x-h)**2/a**2), [1])
    plt.plot(h, k, 'go', label='center')
    if a > b:
        c = (a**2-b**2)**.5
        plt.plot(k+c, h, 'ko', label='focus')
        plt.plot(k-c, h, 'ko', label='focus')
    else:
        c = (b**2-a**2)**.5
        plt.plot(k, h+c, 'ko', label='focus')
        plt.plot(k, h-c, 'ko', label='focus')
    plt.axis('equal')
    plt.xlim(-axis_lim, axis_lim)
    plt.ylim(-axis_lim, axis_lim)
    plt.show()
    
interact(draw_plot, a=(.1, 1.5, .3), b=(.1, 2.5, .3), h=(-2.,2.), k=(-2.,2.))
```

```python
# 13. tangent standard form
def f(x, a, h, k):
    return a*np.tan(x-h) + k

def draw_plot(a, h, k):
    x_lim = 5
    x = list(np.linspace(-x_lim, x_lim, 400))
    y = list(map(lambda elem : f(elem, a, h, k), x))
    y_lim = max(max(y), min(y), key=abs)
    for i in range(int(round(x_lim/np.pi))+1):
        pos_asymptote = np.pi/2+h + i*np.pi
        neg_asymptote = np.pi/2+h - i*np.pi
        if pos_asymptote < x_lim:
            plt.vlines(pos_asymptote, y_lim, -y_lim, linestyles='dashed')
        if neg_asymptote > -x_lim:
            plt.vlines(neg_asymptote, y_lim, -y_lim, linestyles='dashed')
    plt.plot(x, y)
    plt.xlim(-x_lim, x_lim)
    plt.show()
    
interact(draw_plot, a=(1.), h=(-1.), k=(1.))
```

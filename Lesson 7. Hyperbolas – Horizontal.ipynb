The hyperbola that opens left and right and centered at (0, 0) has the equation above.  It has two axes of symmetry, as you can see from the diagram:  The conjugate axis and the transverse axis.  Here are a few other measures to know:

- Vertices:  (a, 0) and (-a, 0)
- Foci:  (c, 0) and (-c, 0) where c = $\sqrt{a^2 + b^2}$

Asymptote lines are used as guidelines in sketching the graph of a hyperbola.  To graph the asymptote lines, form a rectangle by using points (-a, b), (-a, -b), (a, b), and (a, -b).  So the equation of the asymptote line is:
![](https://s3.amazonaws.com/armath-notebook-resources/images/horizontal_hyperbola_3.png)

```python
%matplotlib inline

# load necessary libraries
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import *
from collections import OrderedDict

def draw_plot(a, b, h, k):
    axis_lim = 5
    x = np.linspace(-axis_lim, axis_lim, 400)
    y = np.linspace(-axis_lim, axis_lim, 400)
    asymptote_1 = k + 1.*b/a*(x - h)
    asymptote_2 = k - 1.*b/a*(x - h)
    plt.plot(list(x), list(asymptote_1), dashes=[3, 3], color='orange', label='asymptote')
    plt.plot(list(x), list(asymptote_2), dashes=[3, 3], color='orange', label='asymptote')
    x, y = np.meshgrid(x, y)
    plt.contour(x, y,((x-h)**2/a**2 - (y-k)**2/b**2), [1])
    plt.plot(h+a, k, 'ro', label='vertex')
    plt.plot(h-a, k, 'ro', label='vertex')
    plt.plot(h, k, 'go', label='center')
    c = (a**2+b**2)**.5
    plt.plot(h+c, k, 'ko', label='focus')
    plt.plot(h-c, k, 'ko', label='focus')
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = OrderedDict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys(), numpoints=1)
    plt.grid(axis='y', color='r')
    plt.grid(axis='x', color='g')
    plt.xlim(-axis_lim, axis_lim)
    plt.ylim(-axis_lim, axis_lim)
    plt.show()
    
interact(draw_plot, a=(-1., 3., .3), b=(-1., 3., .3), h=(-2.,2.), k=(-2.,2.))
```
